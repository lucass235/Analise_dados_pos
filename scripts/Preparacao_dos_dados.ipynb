### Limpeza de Dados

#### 1. Tratamento de Valores Ausentes

**Observações:**

    ```python

    # Exemplo de imputação pela média para variáveis numéricas

    ```

````python
data.isnull().sum()
````

````python
data['Ganho de Capital'].fillna(data['Ganho de Capital'].mean(), inplace=True)
````

#### 2. Tratamento de Outliers

**Observações:**
- Outliers foram identificados em variáveis numéricas como Idade, Ganho de Capital, e Perda de Capital, Esses outliers podem distorcer a análise e o desempenho de modelos de machine learning, especialmente os modelos baseados em distâncias, como KNN ou SVM.

**Abordagem:**
- Em vez de remover os outliers diretamente (o que pode resultar na perda de dados importantes), podemos considerar técnicas como:
    - Capping (Capagem): Limitar os valores extremos ao percentil 1% e 99% para evitar que outliers distorçam o modelo.
    - Transformações Logarítmicas: Em casos de distribuições altamente assimétricas, como em Capital de Ganho e Perda de capital, a aplicação de uma transformação logarítmica pode ajudar a suavizar os outliers.

````python
# Função para aplicar capagem nos outliers
def cap_outliers(df, column):
    lower_limit = df[column].quantile(0.01)
    upper_limit = df[column].quantile(0.99)
    df[column] = np.clip(df[column], lower_limit, upper_limit)
    return df

# Aplicando capagem nas variáveis com outliers
data = cap_outliers(data, 'Idade')
data = cap_outliers(data, 'Ganho de Capital')
data = cap_outliers(data, 'Perda de Capital')
````

### Tratamento de Variáveis Categóricas
- O tratamento de variáveis categóricas é uma etapa essencial na preparação de dados para modelos de machine learning. Modelos supervisionados geralmente exigem que as variáveis sejam numéricas, então precisamos converter as variáveis categóricas para um formato adequado.

#### 1. Identificação das Variáveis Categóricas

````python
# identificar todas as variáveis categóricas presentes no conjunto de dados
# Identificando as variáveis categóricas no conjunto de dados
categorical_columns = data.select_dtypes(include=['object']).columns
categorical_columns
````

#### 2. Codificação One-Hot

- A codificação one-hot é apropriada quando as variáveis categóricas não possuem uma ordem intrínseca. Cada categoria é transformada em uma coluna binária (0 ou 1)
- Isso é necessário para variáveis como `Classe de Trabalho`, `Ocupação`, e `País de Origem`, que não possuem uma ordem natural.

````python
data_encoded = pd.get_dummies(data, columns=categorical_columns, drop_first=True)
````

````python
data_encoded.head()
````

#### 3. Codificação de Rótulos (Label Encoding)
- A codificação de rótulos (label encoding) é apropriada quando as variáveis categóricas possuem uma ordem ou um ranking natural. Isso é comum em variáveis como `Educação` (ex.: ensino médio, graduação, pós-graduação), No caso da variável `Classe`, que é binária (<=50K ou >50K), também podemos aplicar a codificação de rótulos.

````python
from sklearn.preprocessing import LabelEncoder

# Aplicando Label Encoding na variável 'income'
le = LabelEncoder()
data_encoded['Classe'] = le.fit_transform(data['Classe'])

# Exibindo as primeiras linhas para verificar o resultado
data_encoded['Classe'].head()
````

### Normalização ou Padronização de Variáveis Numéricas
- A normalização e padronização de variáveis numéricas são técnicas cruciais para preparar dados para algoritmos de machine learning. Esses processos ajustam as escalas das variáveis, garantindo que nenhuma variável numérica tenha uma influência desproporcional no modelo.

#### 1. Decisão de Normalização ou Padronização
**Normalização**
- A normalização transforma os dados para um intervalo específico, como [0, 1]. Isso é útil quando os dados não seguem uma distribuição normal, como em Capital de ganho e Perda de capital, ou quando a escala das variáveis varia significativamente.

**Padronização**
- A padronização transforma os dados para que tenham média zero e desvio padrão igual a um. Isso é recomendado quando as variáveis seguem uma distribuição aproximadamente normal, como Idade e Horas por semana.

#### 2. Implementação da Normalização ou Padronização
- Com base na analise exploratoria feita aplicaremos a padronização para variáveis que têm distribuições mais simétricas e normalizadas como: `Idade`, `Horas por Semana`, `N de Educação`.
- A Normalização será aplicada para variáveis com muitos outliers como `Ganho de Capital`, `Perda de Capital` e `fnlwgt`.

````python
from sklearn.preprocessing import StandardScaler

# Padronização das variáveis numéricas
scaler = StandardScaler()
data[['Idade', 'N da Educação', 'Horas por Semana']] = scaler.fit_transform(data[['Idade', 'N da Educação', 'Horas por Semana']])

# Exibindo as primeiras linhas para verificar o resultado
data[['Idade', 'N da Educação', 'Horas por Semana']].head()
````

````python
from sklearn.preprocessing import MinMaxScaler

# Normalização das variáveis com valores assimétricos
minmax_scaler = MinMaxScaler()
data[['Ganho de Capital', 'Perda de Capital', 'fnlwgt']] = minmax_scaler.fit_transform(data[['Ganho de Capital', 'Perda de Capital', 'fnlwgt']])

# Exibindo as primeiras linhas para verificar o resultado
data[['Ganho de Capital', 'Perda de Capital', 'fnlwgt']].head()
````

#### 3. Razão para a Escolha
- Padronização: Utilizada para variáveis como Idade, N da Educação, e Horas por Semana, que têm distribuições aproximadamente normais. A padronização ajuda esses dados a terem média zero e desvio padrão igual a um, o que é ideal para modelos como regressão linear e SVM.

- Normalização: Utilizada para variáveis como Ganho de Capital, Perda de Capital, e fnlwgt, que possuem distribuições muito assimétricas ou com muitos outliers. A normalização garante que os valores fiquem em um intervalo uniforme, como [0, 1], o que é útil para evitar a influência desproporcional de valores extremos.

### Engenharia de Características
- etapa crítica na preparação de dados para modelos de machine learning. Com base nos insights obtidos durante a análise exploratória, podemos considerar a criação de novas variáveis, transformações e interações entre características existentes para melhorar o desempenho dos modelos.

#### 1. Criação de Novas Caraacterísticas
- Podemos criar uma nova variável que estime o número total de horas trabalhadas ao longo da vida de um indivíduo com base em sua idade e horas trabalhadas por semana.

````python
# Criando uma nova variável 'lifetime_work_hours'
data['Horas por Idade'] = data['Idade'] * data['Horas por Semana']

# Verificando as primeiras linhas da nova variável
data[['Idade', 'Horas por Semana', 'Horas por Idade']].head()
````

#### 2.Interações entre Características
- Podemos criar uma variável que capture a interação entre o nível de educação (em anos) e o número de horas trabalhadas por semana.

````python
# Criando uma nova variável 'education_hours_interaction'
data['Interacao educacao horas'] = data['N da Educação'] * data['Horas por Semana']

# Verificando as primeiras linhas da nova variável
data[['N da Educação', 'Horas por Semana', 'Interacao educacao horas']].head()
````

#### 3. Transformações Logarítmicas
- Variáveis com distribuições altamente assimétricas, como Ganho de Capital e Perda de Capital, podem se beneficiar de transformações logarítmicas para suavizar a distribuição e reduzir a influência de outliers.

````python
# Aplicando transformação logarítmica nas variáveis 'capital-gain' e 'capital-loss'
data['log Ganho de Capital'] = np.log1p(data['Ganho de Capital'])  # log(1 + x) para evitar log(0)
data['log Perda de Capital'] = np.log1p(data['Perda de Capital'])

# Verificando as primeiras linhas das variáveis transformadas
data[['Ganho de Capital', 'log Ganho de Capital', 'Perda de Capital', 'log Perda de Capital']].head()
````
