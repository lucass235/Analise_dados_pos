### Limpeza de Dados

#### 1. Tratamento de Valores Ausentes

**Observações:**
- Na analise feita foram encontrados valores ausentes nas variáveis: `Classe Trabalhadora`, `Ocupação`, `País de Origem`, `Ganho de Capital`, e `Perda de Capital`. 

- Exemplo de uso da tecnica de imputação pela moda para as variáveis categóricas e pela média para as variáveis numéricas:
    ```python	
    # Imputação pela moda para variáveis categóricas
    data['Nome da coluna'].fillna(data['Nome da coluna'].mode()[0], inplace=True)
    # Exemplo de imputação pela média para variáveis numéricas
    data['Nome da coluna'].fillna(data['Nome da coluna'].mean(), inplace=True)
    ```

````python
# For 'Ganho de Capital' and 'Perda de Capital', replace 0 with the mean of the column
data['Ganho de Capital'] = data['Ganho de Capital'].replace(0, data['Ganho de Capital'].mean())
data['Perda de Capital'] = data['Perda de Capital'].replace(0, data['Perda de Capital'].mean())

# For categorical variables with '?', replace with the mode
data['Classe Trabalhadora'] = data['Classe Trabalhadora'].replace(' ?', data['Classe Trabalhadora'].mode()[0])
data['Ocupação'] = data['Ocupação'].replace(' ?', data['Ocupação'].mode()[0])
data['País de Origem'] = data['País de Origem'].replace(' ?', data['País de Origem'].mode()[0])
````

#### 2. Tratamento de Outliers

**Observações:**
- Outliers foram identificados em variáveis numéricas como Idade, Ganho de Capital, e Perda de Capital, Esses outliers podem distorcer a análise e o desempenho de modelos de machine learning, especialmente os modelos baseados em distâncias, como KNN ou SVM.

**Abordagem:**
- Em vez de remover os outliers diretamente (o que pode resultar na perda de dados importantes), podemos considerar técnicas como:
    - Capping (Capagem): Limitar os valores extremos ao percentil 1% e 99% para evitar que outliers distorçam o modelo.

````python
# Função para aplicar capagem nos outliers
def cap_outliers(df, column):
    lower_limit = df[column].quantile(0.01)
    upper_limit = df[column].quantile(0.99)
    df[column] = np.clip(df[column], lower_limit, upper_limit)
    return df

# Aplicando capagem nas variáveis com outliers
data = cap_outliers(data, 'Idade')
data = cap_outliers(data, 'Ganho de Capital')
data = cap_outliers(data, 'Perda de Capital')
````

### Normalização ou Padronização de Variáveis Numéricas
- A normalização e padronização de variáveis numéricas são técnicas cruciais para preparar dados para algoritmos de machine learning. Esses processos ajustam as escalas das variáveis, garantindo que nenhuma variável numérica tenha uma influência desproporcional no modelo.

#### 1. Decisão de Normalização ou Padronização
**Normalização**
- A normalização transforma os dados para um intervalo específico, como [0, 1]. Isso é útil quando os dados não seguem uma distribuição normal, como em Capital de ganho e Perda de capital, ou quando a escala das variáveis varia significativamente.

**Padronização**
- A padronização transforma os dados para que tenham média zero e desvio padrão igual a um. Isso é recomendado quando as variáveis seguem uma distribuição aproximadamente normal, como Idade e Horas por semana.

#### 2. Implementação da Normalização ou Padronização
- Com base na analise exploratoria feita aplicaremos a padronização para variáveis que têm distribuições mais simétricas e normalizadas como: `Idade`, `Horas por Semana`, `N de Educação`.
- A Normalização será aplicada para variáveis com muitos outliers como `Ganho de Capital`, `Perda de Capital` e `fnlwgt`.

````python
# Padronização das variáveis numéricas
scaler = StandardScaler()
data[['Idade', 'N da Educação', 'Horas por Semana']] = scaler.fit_transform(data[['Idade', 'N da Educação', 'Horas por Semana']])

# Exibindo as primeiras linhas para verificar o resultado
data[['Idade', 'N da Educação', 'Horas por Semana']].head()
````

````python
# Normalização das variáveis com valores assimétricos
minmax_scaler = MinMaxScaler()
data[['Ganho de Capital', 'Perda de Capital', 'fnlwgt']] = minmax_scaler.fit_transform(data[['Ganho de Capital', 'Perda de Capital', 'fnlwgt']])

# Exibindo as primeiras linhas para verificar o resultado
data[['Ganho de Capital', 'Perda de Capital', 'fnlwgt']].head()
````

#### 3. Razão para a Escolha
- Padronização: Utilizada para variáveis como Idade, N da Educação, e Horas por Semana, que têm distribuições aproximadamente normais. A padronização ajuda esses dados a terem média zero e desvio padrão igual a um, o que é ideal para modelos como regressão linear e SVM.

- Normalização: Utilizada para variáveis como Ganho de Capital, Perda de Capital, e fnlwgt, que possuem distribuições muito assimétricas ou com muitos outliers. A normalização garante que os valores fiquem em um intervalo uniforme, como [0, 1], o que é útil para evitar a influência desproporcional de valores extremos.

### Balanceamento de Classes (Bônus)
- Em problemas de classificação, é comum encontrar classes desbalanceadas, o que pode levar a modelos que favorecem a classe majoritária e têm um desempenho ruim na classe minoritária. No conjunto de dados "Adult Income", a variável-alvo income é uma variável binária que classifica indivíduos com rendimentos <=50K e >50K. Dependendo da distribuição dessas classes, pode ser necessário aplicar técnicas de balanceamento para garantir que o modelo tenha uma performance adequada em ambas as classes.

````python
# Verificando a distribuição das classes na variável 'income'
data['Classe'].value_counts(normalize=True)
````

#### Técnicas de Balanceamento de Classes
- **Sobreamostragem da Classe Minoritária (Oversampling)**: Sobreamostragem consiste em duplicar ou gerar exemplos sintéticos da classe minoritária para equilibrar a distribuição entre as classes.

````python
# Técnica: SMOTE (Synthetic Minority Over-sampling Technique) é uma técnica popular que cria exemplos sintéticos da classe minoritária.
from imblearn.over_sampling import SMOTE

# Supondo que 'data' é o seu DataFrame e 'Classe' é a variável alvo

# Identificar colunas categóricas (exemplo)
colunas_categoricas = data.select_dtypes(include=['object']).columns

# Aplicar One-Hot Encoding nas variáveis categóricas
data_codificado = pd.get_dummies(data, columns=colunas_categoricas)

# Definindo X e y para o balanceamento após codificação
X = data_codificado.drop('Classe', axis=1)
y = data_codificado['Classe']

# Aplicando SMOTE
smote = SMOTE(random_state=42)
X_smote, y_smote = smote.fit_resample(X, y)

# Verificando a nova distribuição das classes
print(pd.Series(y_smote).value_counts(normalize=True))
````

- **Subamostragem da Classe Majoritária (Undersampling)**: Subamostragem remove exemplos da classe majoritária para igualar a distribuição de classes.

````python
# A subamostragem simples reduz o número de exemplos da classe majoritária para se alinhar com a classe minoritária.
from imblearn.under_sampling import RandomUnderSampler

# Aplicando subamostragem na classe majoritária
rus = RandomUnderSampler(random_state=42)
X_rus, y_rus = rus.fit_resample(X, y)

# Verificando a nova distribuição das classes
pd.Series(y_rus).value_counts(normalize=True)
````

- **Métodos Baseados em Ponderação (Weighted Methods)**: Em vez de alterar a distribuição das classes, os métodos de ponderação ajustam a importância das classes minoritárias atribuindo pesos maiores a elas durante o treinamento do modelo.

````python
# Algoritmos como Random Forest e SVM permitem atribuir pesos diferentes para as classes, dando mais importância à classe minoritária.
from sklearn.ensemble import RandomForestClassifier

# Criando um modelo de Random Forest com pesos ajustados
rf_model = RandomForestClassifier(class_weight='balanced', random_state=42)

# Treinando o modelo
rf_model.fit(X, y)
````

#### Escolha da Técnica
**Sobreamostragem**
-  Indicada para algoritmos que se beneficiam de mais dados, como redes neurais e regressão logística.

**Subamostragem**
-  Adequada para modelos que podem processar menos dados de forma eficiente, como árvores de decisão.

**Ponderação de Classes**
- Útil para modelos que permitem o ajuste de pesos de classes, como Random Forest e SVM.
